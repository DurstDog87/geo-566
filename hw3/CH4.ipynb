```python
#85, Hypotenuse of a triangle
from math import sqrt
def hypotenuse(a: float=1, b: float=1) -> float:
    return sqrt(a**2 + b**2)

def main():
    print("Enter the short sides of your triangle")
    hyp = hypotenuse(float(input("a: ")), float(input("b: ")))
    print(f"The hypotenuse is: {hyp}")

if __name__ == "__main__" : main()
```

```python
#86, Taxi Fare

#define current taxi rates
rates = {           
    "base" : 4.00,
    "rate" : 0.25
}
def calculateFare(dist: float, rates: dict) -> float:
    return (((dist*1000)/140)*rates["rate"]) + rates["base"] #calculate fare based on current rates
#main function
def main():
    print("Taxi fare calculator")
    dist = float(input("How many kilometers was your ride? "))
    fare = calculateFare(dist, rates)
    print(f"Your taxi fare for {dist}km is ${fare}")

if __name__ == "__main__" : main()
    
```

```python
#87, Shipping calculator

#define shipping rates
rates = {
    "1" : 10.95,
    "1+" : 2.95
}
def shippingCharge(nitems: int, rate: dict=rates) -> float:
    return rate["1"] + (rate["1+"]*(nitems-1)) if nitems > 0 else 0 #calculate shipping rate, return 0 if no items are added

#main function
def main():
    nitems = int(input("Enter number of items"))
    print(f"Your shipping charge for {nitems} items is ${shippingCharge(nitems)}")

if __name__ == "__main__" : main()
```

```python
#88, Median of 3 values

#This function can find the median of any list of numbers
def median(numbers: list) -> float:
    numbers.sort() #sort the list ascending 
    med = int(len(numbers)/2) #find the middle
    return (numbers[med] + numbers[med+1])/2 if (len(numbers))%2 == 0 else numbers[med] #return the middle number
     

def main():
    numbers = list(input("enter a string of digits: "))
    numbers = [float(i) for i in numbers if i.isdigit()] #get only digits from the string
    print(f"The median is: {median(numbers)}") #print output to the console
    
if __name__ == "__main__" : main()
```

```python
#91, Gregorian date to ordinal date
from calendar import monthrange
def ordinalDate(d: int=1, m: int=0, y: int=2000) -> int:
    orDay = sum([monthrange(y, i+1)[1] for i in range(m-1)]) + d #sum of total days for each full month add day of the month
    if orDay > 365:                                              #if the day > 365, iterate again, turn over the year
        return ordinalDate(y=y+1, m=0, d=orDay-365)
    else:                                                        #otherwise just return day and year
        return orDay, y

def main():
    print("Gregorian date to ordinal date")
    m = int(input("Enter the month: "))
    d = int(input("Enter the day: "))
    y = int(input("Enter the year: "))
    day, year = ordinalDate(d,m,y)
    print(f"The ordinal date is day {day} of {year}")

if __name__ == "__main__" : main()
```

```python
"""
At the end of this page is the code for a contract that was in the presentation.  Modify it as follows:
	•	Add an end date as one of the class member values.
	•	Add a method that will calculate the number of days from the start to the end of the contract.  (Take time to research how such a calculation is done.)
	•	Create three instances of using the revised class similar to what was done in the sample code provided.  
    For each instance, there should be a single line report that reflects the parties involve, 
    the purpose of the contract (e.g. a rental), the start and end date of the contract and the total number of days for the contract duration.
	•	Place all the code in the same Jupyter file from Part I of the assignment.
"""

from datetime import datetime, date
from calendar import monthrange

#this is the function defined for problem 91
def ordinalDate(d: int=1, m: int=0, y: int=2000) -> int:
        orDay = sum([monthrange(y, i+1)[1] for i in range(m-1)]) + d #sum of total days for each full month add day of the month
        if orDay > 365:                                              #if the day > 365, iterate again, turn over the year
            return ordinalDate(y=y+1, m=0, d=orDay-365)
        else:                                                        #otherwise just return day and year
            return orDay, y


class Contract:
    def __init__(self,seller,buyer,item,endDate):
        self.seller = seller
        self.buyer = buyer
        self.cdate = date.today()
        self.item = item
        self.end = endDate


    def contractLength(self) -> int:
        parse = lambda s, d : str(s).split(d) #define helper to parse date strings
        dlim = "-"
        start, end = [int(i) for i in parse(self.cdate,dlim)], [int(i) for i in parse(self.end,dlim)] #parse inputs
        startD, startY  = ordinalDate(y=start[0],m=start[1],d=start[2])   # get the ordinal dates
        endD, endY = ordinalDate(y=end[0],m=end[1],d=end[2]) 
        return (endD-startD) + ((endY-startY)*365)                        #return contract length in days

    def details(self):
        print (self.seller + " sells " 
               + self.item + " to " + self.buyer
               + "\nContract start: " + str(self.cdate)
               + "\nContract end: " + self.end
               + "\nContract length: " + str(self.contractLength()) + " days\n")
        
    @staticmethod
    def laborcost(hours, wage):
        return hours * wage


def main(): #test
    c1 = Contract("Joe", "Pat", "car", "2022-01-30")
    c2 = Contract("Mary,", "Pat", "house", "2023-1-26")
    c3 = Contract("Joe", "Jacinda", "1 single bean", str(datetime.strptime('Jun 25 2025', '%b %d %Y')).split(' ')[0])
    c1.details()
    c2.details()
    c3.details()
    print(f"The labor cost for this week is ${Contract.laborcost(40,15.00)}")

if __name__ == "__main__" : main()
```
